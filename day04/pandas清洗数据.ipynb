```python
import pandas as pd
import numpy as np
```

```python
# 读取两张表
df = pd.read_excel('./data/E Commerce Dataset.xlsx',sheet_name=None)

df1 = df['Data Dict']
df2 = df['E Comm']
df2.head(10)
```

### 1.输出用户数据每个字段的缺失数量和缺失比例

```python
# 1.输出缺失数量
mis_count = df2.isnull().sum().sort_values(ascending=False)
print(mis_count)
# 2.输出缺失比例
mis_present = (df2.isnull().mean()*100).round(1).sort_values(ascending=False)
print(mis_present)
```

### 2.用中位数填补用户数据中的数值缺失值；

```python
print(df2.info())
print(df2.isnull())
# 使用for循环的方式，使用每列的中位数 来填充各列的缺失值
# 1.获取每个列名
for col_name in df2.columns:
    # 2.判断该列是否含有缺失值
    if np.all(pd.notnull(df2[col_name])) == False and df2[col_name].dtype in ['int64', 'float64', 'int32', 'float32']:
        # 3.到这里说明有缺失值
        print(col_name)     # 打印列名
        # 4.打印中位数
        print(df2[col_name].median())
        # 5.用该列的平均值来填充该列的缺失值
        df2[col_name] = df2[col_name].fillna(df2[col_name].median())
```
