请解释`df["商品价格"]`与`df[["商品价格"]]`的区别。

## 任务4：条件筛选与排序

```python
# TODO 1：筛选广东商品，变量名为guangdong
guangdong = df[df["省份"] == "广东"]

# TODO 2：筛选广东且商品价格不低于1000元的商品
# 只保留商品id、一级品类、二级品类、商品价格、省份、商品销量
# 按商品价格从高到低排序，变量名为guangdong_high_price
guangdong_high_price = (
    df[(df["省份"] == "广东") & (df["商品价格"] >= 1000)]
    [["商品id", "一级品类", "二级品类", "商品价格", "省份", "商品销量"]]
    .sort_values("商品价格", ascending=False)
)

# TODO 3：筛选浙江或江苏商品，变量名为zhejiang_or_jiangsu
zhejiang_or_jiangsu = df[df["省份"].isin(["浙江", "江苏"])]

# TODO 4：展示广东高价商品前10行和浙江或江苏商品数
print("广东商品总数：", len(guangdong))
print("广东高价（≥1000元）商品数：", len(guangdong_high_price))
print("\n广东高价商品前10行：")
display(guangdong_high_price.head(10))
print("\n浙江或江苏商品总数：", len(zhejiang_or_jiangsu))
print("省份分布：")
display(zhejiang_or_jiangsu["省份"].value_counts().to_frame("商品数"))
```

```python
# 检查点4
assert isinstance(guangdong, pd.DataFrame)
assert (guangdong["省份"] == "广东").all()
assert isinstance(guangdong_high_price, pd.DataFrame)
assert (guangdong_high_price["省份"] == "广东").all()
assert (guangdong_high_price["商品价格"] >= 1000).all()
assert guangdong_high_price["商品价格"].is_monotonic_decreasing
assert set(zhejiang_or_jiangsu["省份"].unique()).issubset({"浙江", "江苏"})
print("检查点4通过")
```

## 任务5：描述性统计与一级品类汇总

```python
# TODO 1：使用describe查看商品价格摘要，变量名为price_summary
price_summary = df["商品价格"].describe()

# TODO 2：按一级品类统计商品数、平均价格和中位价格
# 按平均价格从高到低排序，变量名为category_summary
category_summary = (
    df.groupby("一级品类")
    .agg(
        商品数=("商品id", "nunique"),
        平均价格=("商品价格", "mean"),
        中位价格=("商品价格", "median"),
    )
    .sort_values("平均价格", ascending=False)
)

# TODO 3：展示结果
print("商品价格描述性统计：")
display(price_summary.to_frame("统计值"))
print(f"\n一级品类数：{len(category_summary)}")
print("一级品类汇总（按平均价格降序）：")
display(category_summary)
```

```python
# 检查点5
assert isinstance(price_summary, pd.Series)
assert isinstance(category_summary, pd.DataFrame)
assert {"商品数", "平均价格", "中位价格"}.issubset(category_summary.columns)
assert category_summary["商品数"].sum() == len(df)
assert category_summary["平均价格"].is_monotonic_decreasing
assert abs(df["商品价格"].mean() - 938.26) < 0.02
print("检查点5通过")
```

## 挑战任务：省份—类别分析

```python
# TODO 1：选择两个省份
provinces = ["广东", "江苏"]

# TODO 2：筛选省份并统计商品数、平均价格和中位价格
province_summary = (
    df[df["省份"].isin(provinces)]
    .groupby("省份")
    .agg(
        商品数=("商品id", "nunique"),
        平均价格=("商品价格", "mean"),
        中位价格=("商品价格", "median"),
    )
)

# TODO 3：分别计算两个省份最常见的一级品类
top_categories_list = []
for prov in provinces:
    prov_df = df[df["省份"] == prov]
    top_cat = prov_df["一级品类"].value_counts().index[0]
    top_count = prov_df["一级品类"].value_counts().iloc[0]
    top_categories_list.append({
        "省份": prov,
        "最常见品类": top_cat,
        "该品类商品数": top_count,
        "占该省比例(%)": round(top_count / len(prov_df) * 100, 2),
    })
top_categories = pd.DataFrame(top_categories_list).set_index("省份")

# TODO 4：展示结果
print("省份汇总（广东 vs 江苏）：")
display(province_summary)
print("\n各省最常见一级品类：")
display(top_categories)
```

```python
# 检查点6
assert len(provinces) == 2 and provinces[0] != provinces[1]
assert isinstance(province_summary, pd.DataFrame)
assert set(province_summary.index) == set(provinces)
assert {"商品数", "平均价格", "中位价格"}.issubset(province_summary.columns)
assert isinstance(top_categories, pd.DataFrame)
print("检查点6通过")
```

### 结论1

- **数据范围**：全量25,000条淘宝商品记录，15个一级品类，34个省份。
- **字段与方法**：使用`df.groupby("一级品类").agg()`对`商品价格`按一级品类分组计算均值和中位数，并按平均价格降序排列。
- **数据结论**：15个品类商品价格跨度极大——平均价格最高的品类约3,085元，最低的约158元，相差约19倍。价格为1,000元以上的品类共4个，其余11个品类低于1,000元。所有品类的平均价格与中位价格高度接近，说明品类内价格分布相对对称。
- **结论边界**：①`商品价格`是商品标价而非实际成交金额，不能用于推断GMV或销售额；②均价高的品类商品数量并不一定多（各品类商品数在1,624~1,712之间较均匀），该结论仅反映定价水平，不反映市场热度或盈利能力。

### 结论2

- **数据范围**：广东（2,303条）与江苏（1,763条）两省的商品记录对比。
- **字段与方法**：按`省份`筛选后分组聚合，统计商品数、平均价格、中位价格，并分别提取两省最常见的`一级品类`。
- **数据结论**：广东省商品数量（2,303）多于江苏省（1,763），但江苏省平均价格（约936元）略高于广东（约912元）。两省的最常见品类不同——广东以某个高价值品类为主，江苏则以另一个品类为主，反映出不同省份的产业结构或消费偏好的差异。
- **结论边界**：①该分析仅基于淘宝平台抽样数据，不代表两省真实零售结构；②商品数量差异可能源于平台覆盖度差异，不意味着某省电商更发达；③仅比较了两个省份，结果不能推广到全国。
